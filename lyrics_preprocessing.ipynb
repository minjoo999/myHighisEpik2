# 중복 제거

```python
du = df.duplicated(['제목'], keep=False)
dfdu = pd.concat([df, du], axis=1)
dfdu.rename(columns={0:'중복여부'}, inplace=True)
dfdu
```

```python
non = dfdu[(dfdu['중복여부'] == False)]
```

```python
dup = dfdu[(dfdu['중복여부'] == True)]
dup
```

```python
len(dup)
```

```python
dup.drop(dup[(dup['앨범'] == "에픽하이 3집 리패키지 - Black Swan Songs") | (dup['앨범'] == "Breakdown")].index, inplace=True)
```

```python
dup
```

```python
len(dup)
```

```python
dup.drop(dup[(dup['앨범'] == "비 오는 날 듣기 좋은 노래") | (dup['앨범'] == "달의 연인 - 보보경심 려 OST Part 6 (SBS 월화드라마)") | (dup['앨범'] == "논스톱 5 OST") | (dup['앨범'] == "New K-Pop Compilation")].index, inplace=True)
dup
```

```python
# non + dup
df_all = pd.concat([non, dup], ignore_index=True)
df_all
```

```python
df_all.to_csv('tracks_final.csv', sep=',', index=False)
```

# 가사 전처리

```python
lyrics = ' '.join(df_all['가사'])
lyrics
```

```python
lyrics2 = lyrics.replace("\r\n", " ")
lyrics2
```
